```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.stats.proportion import power_proportions_2indep, test_proportions_2indep
```

```python
ab = pd.read_csv('ab.csv')
```

```python
ab.head()
```

```python
def analysis(ab, start_date, end_date):
    result = pd.DataFrame(columns=[
        'start_date',
        'end_date',
        'daily_visit_control',
        'agg_visit_control',
        'daily_visit_treatment',
        'agg_visit_treatment',
        'daily_cr_control',
        'agg_cr_control',
        'daily_cr_treatment',
        'agg_cr_treatment',
        'daily_power',
        'agg_power',
        'daily_p_value',
        'agg_p_value'
    ])
    
    daterange = pd.date_range(start_date, end_date)
    
    for date in daterange:
        true_end_date = str(date + pd.DateOffset(1))
        
        daily = ab[
            (ab['timestamp'] >= str(date)) &
            (ab['timestamp'] <= true_end_date)
        ]
        
        agg = ab[
            (ab['timestamp'] >= start_date) &
            (ab['timestamp'] <= true_end_date)
        ]
        
        daily = daily.groupby(ab.group).agg(
            start_date=pd.NamedAgg(column='timestamp', aggfunc=lambda x: min(x).split(' ')[0]),
            end_date=pd.NamedAgg(column='timestamp', aggfunc=lambda x: max(x).split(' ')[0]),
            visit=pd.NamedAgg(column='converted', aggfunc=len),
            converted=pd.NamedAgg(column='converted', aggfunc=sum),
            convention_rate=pd.NamedAgg(column='converted', aggfunc=lambda x: sum(x) / len(x))
        )
        
        agg = agg.groupby(ab.group).agg(
            start_date=pd.NamedAgg(column='timestamp', aggfunc=lambda x: min(x).split(' ')[0]),
            end_date=pd.NamedAgg(column='timestamp', aggfunc=lambda x: max(x).split(' ')[0]),
            visit=pd.NamedAgg(column='converted', aggfunc=len),
            converted=pd.NamedAgg(column='converted', aggfunc=sum),
            convention_rate=pd.NamedAgg(column='converted', aggfunc=lambda x: sum(x) / len(x))
        )
        
        daily_power_proportions = power_proportions_2indep(
            diff=daily['convention_rate']['control'] - daily['convention_rate']['treatment'],
            prop2=daily['convention_rate']['treatment'],
            nobs1=daily['visit']['control'],
            ratio=daily['visit']['treatment'] / daily['visit']['control'],
            alpha=0.05,
            alternative='two-sided',
            return_results=True
        )
        
        agg_power_proportions = power_proportions_2indep(
            diff=agg['convention_rate']['control'] - agg['convention_rate']['treatment'],
            prop2=agg['convention_rate']['treatment'],
            nobs1=agg['visit']['control'],
            ratio=agg['visit']['treatment'] / agg['visit']['control'],
            alpha=0.05,
            alternative='two-sided',
            return_results=True
        )
                    
        daily_analysis = test_proportions_2indep(
            count1=daily['converted']['control'], 
            nobs1=daily['visit']['control'], 
            count2=daily['converted']['treatment'], 
            nobs2=daily['visit']['treatment'], 
            alternative='two-sided'
        )
        
        agg_analysis = test_proportions_2indep(
            count1=agg['converted']['control'], 
            nobs1=agg['visit']['control'], 
            count2=agg['converted']['treatment'], 
            nobs2=agg['visit']['treatment'], 
            alternative='two-sided'
        )
        
        res = {
            'start_date': start_date,
            'end_date': date,
            'daily_visit_control': daily['visit']['control'],
            'agg_visit_control': agg['visit']['control'],
            'daily_visit_treatment': daily['visit']['treatment'],
            'agg_visit_treatment': agg['visit']['treatment'],
            'daily_cr_control': daily['convention_rate']['control'],
            'agg_cr_control': agg['convention_rate']['control'],
            'daily_cr_treatment': daily['convention_rate']['treatment'],
            'agg_cr_treatment': agg['convention_rate']['treatment'],
            'daily_power': daily_power_proportions.power,
            'agg_power': agg_power_proportions.power,
            'daily_p_value': daily_analysis[1],
            'agg_p_value': agg_analysis[1]
        }

        result = result.append(res, ignore_index=True)
    
    return result
```

```python
start_date = '2017-01-02'
end_date = '2017-01-24'

result = analysis(ab, start_date, end_date)
result.head()
```

```python
plt.figure(figsize=(10, 5), facecolor='white')

plt.plot(np.arange(2,25), result['agg_p_value'])
plt.plot(np.arange(2,25), result['agg_power'])

plt.show()
```
