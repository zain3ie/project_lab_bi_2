```python
import datetime
import pandas as pd
import numpy as np
```

```python
order_items = pd.read_csv('olist_order_items_dataset.csv')
```

```python
order_items['shipping_limit_date'] = order_items['shipping_limit_date'].str[:7]
```

```python
order_items
```

```python
revenue = order_items.sort_values('shipping_limit_date').groupby('shipping_limit_date').agg(
    revenue=pd.NamedAgg(column='price', aggfunc=sum),
    total_order=pd.NamedAgg(column='order_id', aggfunc=lambda x: x.nunique()),
)
```
